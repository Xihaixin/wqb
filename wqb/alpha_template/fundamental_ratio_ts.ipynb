time_series_operator(<profit_field>/<size_field>, <days>)

`implementation`

- 使用时间序列运算符和两个基本指标的比率
- profit_field 代表营收/利润/盈利等相关字段
- size_field 则表示公司规模
- days 使用合理的日期参数

# 3. Alpha Factory

通过构造一个alpha factory来批量生成对应的alpha表达式。传入对应的字段和

````python
size_fields = ["fnd17_mktcap", "shareout"]

## ts_factory在machine_lib里面有定义
# def ts_factory(op, field):
#     output = []
#     #days = [3, 5, 10, 20, 60, 120, 240]
#     days = [5, 22, 66, 120, 240]
    
#     for day in days:
    
#         alpha = "%s(%s, %d)"%(op, field, day)
#         output.append(alpha)
    
#     return output

def fundamental_factory(profit_fields, size_fields, ops):
    alpha_list = []
    for profit_field in profit_fields:
        for size_field in size_fields:
            base_data = "%s/%s" % (profit_field, size_field)
            for ts_op in ts_ops:
                alpha_list += ts_factory(ts_op, base_data)
    return alpha_list

alpha_list = fundamental_factory(profit_fields, size_fields, ts_ops)
for alpha in alpha_list:
    print(alpha)

print("数量: %s"%len(alpha_list))
````

# 7.剪枝操作

原剪枝操作有一些地方考虑得不太全面。它将所有在给定日期内满足条件的alpha筛选出来，但是这些alpha并不一定是我们的策略所构造的alpha。此外，它通过简单的字符串分割并不一定能得到我们想要的字段，有可能会混入多余的字符串。

剪枝的目的是什么？
筛选出我们正在研究的数据字段的alpha表达式

1. 应该添加检查逻辑 - 确保只有当prefix确实存在表达式中时，才进行字段提取
2. 提取逻辑存在风险
    - 前缀匹配不够精确 -> 使用正则表达式
    - 逗号分割不可靠 -
    - 边界情况处理不足
```python
def prune(next_alpha_recs, prefix, keep_num):
    # prefix is the datafield prefix, fnd6, mdl175 ...
    # keep_num is the num of top sharpe same-datafield alpha
    output = []
    num_dict = defaultdict(int)
    for rec in next_alpha_recs:
        exp = rec[1]
        field = exp.split(prefix)[-1].split(",")[0]
        sharpe = rec[2]
        if sharpe < 0:
            field = "-%s"%field
        if num_dict[field] < keep_num:
            num_dict[field] += 1
            decay = rec[-1]
            exp = rec[1]
            output.append([exp,decay])
    return output

```

````python

def prune_improved(next_alpha_recs, contains):
    output = []
    for rec in next_alpha_recs:
        exp = rec[1]
        decay = rec[-1]
        # 检查是否包含所有指定的字符串
        for contain in contains:
            if contain in exp: continue
        output.append([exp, decay])
    return output
````
