# [Alpha Template] Time Series Sentiment comparision

**标题含义：时间序列舆情比较的模板**

`idea`

假设：当一家公司正向舆情信号减去负面舆情信号为正时，我们就估计该公司的股价会拉升。

`template`

<time_series_operator>(<positive_sentiment> - <negative_sentiment>, <days>)

`implementation`

- 使用时间序列运算符和两个基本指标的差值
- days 使用合理的日期参数
- positive_sentiment 正向风评
- negative_sentiment 负向风评

扩展：

1. 使用rank对两个字段进行分别处理
2. 比较两个处理后的数据字段 compare

还能够使用 news 数据集

此外，对于正向和负向舆情数据，似乎并不能直接简单地将它们两两配对。可以观察到

```python
positive_fields = df[df["id"].str.contains("pos")]["id"].values.to_list()
negative_fields = df[df["id"].str.contains("neg")]["id"].values.to_list()

print(positive_fields)
print(negative_fields)
```

```python
def sentiment_factory(pos_fields, neg_fields, ts_ops):
    "<time_series_operator>(<positive_sentiment> - <negative_sentiment>, <days>)"
    alpha_expression = []
    days = [3, 5, 10, 20, 60, 120, 240]
    for ts_opt in ts_ops:
        for pos_field in pos_fields:
            for neg_field in neg_fields:
                for day in days:
                    alpha = "ts_opt(%s - %s, %d)" %(pos_field, neg_field, day)
                    alpha_expression.append(alpha)                
    return alpha_expression

    
```
